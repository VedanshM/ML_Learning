```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('dark_background')
```

```
y1 = np.array(list(range(10)))
y2 = y1**2
y3 = y1**3
```

```
x = y1
```

```
fig, ax = plt.subplots(1, 3)
p,q,r = ax
p.plot(x, y1)
p.set_title('linear')
q.plot(x, y2)
q.set_title('quad')
r.plot(x, y3)
r.set_title('cubic')
fig.tight_layout()
plt.show()
```

```
historyDict = {'loss': [1,2,3,4], 'val_loss':[1,3,9,27], 'accuracy':[9,8,7,6], 'val_accuracy':[5,4,3,2,1]}
plt.plot(historyDict['loss'])
plt.title('training loss')
plt.show()
plt.plot(historyDict['val_loss'])
plt.title('val loss')
plt.show()
plt.plot(historyDict['accuracy'], label='training')
plt.plot(historyDict['val_accuracy'], label='val')
plt.legend()
plt.title('accuracy')
plt.show()
```
